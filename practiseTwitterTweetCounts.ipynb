```python
import pip._vendor.requests 
import os
import json
from dotenv import load_dotenv

load_dotenv()

# To set your environment variables in your terminal run the following line:
# export 'BEARER_TOKEN'='<your_bearer_token>'
bearer_token = os.environ.get("BEARER-TOKEN")

search_url = "https://api.twitter.com/2/tweets/counts/all"

# Optional params: start_time,end_time,since_id,until_id,next_token,granularity
# query_params = {'query': 'china lang:en -is:retweet OR covid lang:en -is:retweet','granularity': 'day', 'start_time': '2021-01-01T00:00:00Z', 'end_time': '2021-03-01T00:00:00Z', 'next_token': {}}
def query_params(next_token):
    return {'query': 'china lang:en -is:retweet OR covid lang:en -is:retweet','granularity': 'day', 'start_time': '2021-01-01T00:00:00Z', 'end_time': '2021-03-01T00:00:00Z', 'next_token': next_token}


def bearer_oauth(r):
    """
    Method required by bearer token authentication.
    """

    r.headers["Authorization"] = f"Bearer {bearer_token}"
    r.headers["User-Agent"] = "v2FullArchiveTweetCountsPython"
    return r


def connect_to_endpoint(url, params):
    response = pip._vendor.requests.request("GET", search_url, auth=bearer_oauth, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()


# def main():
#     json_response = connect_to_endpoint(search_url, query_params)
#     print(json.dumps(json_response, indent=4, sort_keys=True))

def main():
    next_token = {}
    finished = False
    while finished == False:
        json_response = connect_to_endpoint(search_url, query_params(next_token))
        print(json.dumps(json_response, indent=4, sort_keys=True))
        if 'next_token' in json_response['meta']:
            # Save the token to use for next call
            next_token = json_response['meta']['next_token']
        else:
            finished = True


if __name__ == "__main__":
    main()
```
